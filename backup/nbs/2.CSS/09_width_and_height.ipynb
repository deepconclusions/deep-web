## Width and Height

The `width` and `height` CSS properties set the width/heght of an element.

By default, these properties define the width/height of the content area/box

Valid values:

- `<length>` 
- `<percentage>`
- `auto`: calculates and selects a width/height for the element
- `max-content`: the intrinsic prefered width
- `min-content`: the intrinsic minimum height
- `available`: containing block width minus horizontal margin, border and padding
- `fit-content`: the larger of: the intrinsic minimum width; the smaller of the intrinsic preferred width and the available width

<div class="alert text-white rounded" style="background: #3a6e68;"><h4>Note!</h4><p>The `width` and `height` attributes only set the area inside the padding, border and margin of the element anddoes not include them.</p></div>

**Example:**

Here is an example of the `width` CSS property

```css
#div1 {
    width: 200px;
}

#div2 {
    width: 5em;
}

#div3 {
    width: 75%;
}

#div4 {
    width: auto;
}

div {
    height: 200px;
    border: 2px solid red;
    margin: 5px;
}
```

**Result:**

<div style="width: 200px; height: 50px; border: 2px solid red; margin: 5px;"></div>
<div style="width: 5em; height: 50px; border: 2px solid red; margin: 5px;"></div>
<div style="width: 75%; height: 50px; border: 2px solid red; margin: 5px;"></div>
<div style="width: auto; height: 50px; border: 2px solid red; margin: 5px;"></div>


**Example:**

Here is an example of the `height` CSS property

```css
#div1 {
    height: 200px;
}

#div2 {
    height: 5em;
}

#div3 {
    height: 75%;
}

#div4 {
    height: auto;
}

div {
    height: 200px;
    border: 2px solid red;
    margin: 5px;
}
```

**Result:**

<div style="height: 200px; width: 100px; border: 2px solid red; margin: 5px;"></div>
<div style="height: 5em; width: 100px; border: 2px solid red; margin: 5px;"></div>
<div style="height: 75%; width: 100px; border: 2px solid red; margin: 5px;"></div>
<div style="height: auto; width: 100px; border: 2px solid red; margin: 5px;"></div>

## Minimum and Maximum Widths and Heights

The following properties define CSS minimum and maximum widths and heights

- `min-width`: sets the minimum width of an element.
- `max-width`: sets the maximum width of an element.
- `min-height`: sets the minimum height of an element.
- `max-height`: sets the maximum height of an element

Valid values:

- `<length>`
- `<percentage>`

### Examples

Here's an example of the `min-width` CSS property.

```css
p {
    min-width: 500px;
    background: lightblue;
}
```

**Result:**

<p style="min-width: 500px;background: lightblue;">Here's an example of the min-width CSS property.</p>

Here's an example of the `max-width` CSS property.

```css
p {
    max-width: 250px;
    background: lightgreen;
}
```

**Result:**

<p style="max-width: 250px;background: lightgreen;">Here's an example of the max-width CSS property.</p>
